# **Example Use Case: Machine Predictive Maintenance**
---
## Introduction

This dataset contains data from a drilling machine. It consists of 10.000 data points with 13 columns. The first 8 contain sensor measurements of the machine operation.

1. **ID** [s]: Product identifier ranging from 1 to 10000.
2. **product type**: Different product types categorized with A(20%), B(30%), or C(50%).
3. **air temperature** [K]: Ambient temperature measured around the drilling tool.
4. **process temperature** [K]: Temperature of the drilling tool.
5. **torque** [Nm]: Torque applied on the tool.
6. **rotational speed** [rpm]: Measured rotational Speed of the tool.
7. **tool wear** [min]: The amount of time the drill was used.
8. **machine failure** [0/1]: Indicates, whether the machine has failed for any of the following failure modes are true.

The machine failures consist of five independent failure modes

9. **tool wear failure** (TWF): The machine tool was replaced due to tool wear.
10. **heat dissipation failure** (HDF): The heat dissipation is not sufficient which results in overheating.
11. **overstrain failure** (OSF): The wear between workpiece and machine tool results in tool damage.
12. **power failure** (PWF): Occurs if there is a mismatch between provided power and applied power of the process.
13. **random failures** (RNF): Failures which could not be classified to the previous listed.

In the following we will use this dataset and illustrate every step of the data preparation pipeline.

## 4. (Optional / Advanced) Outlier Removal with IQR
Finding the position of the outliers with the Interquartile Range (IQR). It is used to build boxplot graphs. IQR is a concept in statistics that is used to measure the statistical dispersion and data variability by dividing the dataset into quartiles.

1. Any dataset is divided into four defined intervals based upon the values of the data and how they compare to the entire dataset. A quartile is what divides the data into three points and four intervals.
2. Interquartile Range (IQR) is the difference between the third quartile and the first quartile (IQR = Q3 -Q1). Outliers in this case are defined as the observations that are below (Q1 − 1.5x IQR) or boxplot lower whisker or above (Q3 + 1.5x IQR) or boxplot upper whisker.


**Task:** *Write a function that given a dataframe removes all outliers using the IQR method. Test your function on the provided data.*

````python
import pandas as pd

def remove_outliers_iqr(df, columns=None, factor=1.5):
    # Select numeric columns by default
    if columns is None:
        raise ValueError("

    df_clean = df.copy()
    # For each column, compute bounds and filter
    for col in columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower, upper = Q1 - factor * IQR, Q3 + factor * IQR
        df_clean = df_clean[(df_clean[col] >= lower) & (df_clean[col] <= upper)]

    return df_clean
````
